```python
"""
Given an array, check whether the sum of any two numbers in the array
equals a target sum. If yes, output 1, and if not, output 0.
"""
list =[1,2,4,8,16,32,128]
target_sum = 3
length = len(list)
output = 0
for i in range(0,length,1):
    for j in range(i+1,length,1):
        if list[i] + list[j] == target_sum:
            output = 1
            break
    if output == 1:
        break
    
print(output)
```

```python
"""Print the Fibonacci series up to a pre-defined n number.
"""
prenum = 6


def Fibonacci(number):
    if number == 0:
        return 0
    if number == 1 or number == 2:
        return 1
    else:
        return Fibonacci(number-1)+Fibonacci(number-2)
          
for i in range(prenum):
    print(Fibonacci(i), end =" ")
```

Say you have an array with increasing and decreasing sequences. Find
the highest peak in that array. For example, in the array [1, 2, 3, 2, 0], the
third element is the highest in value, i.e., the peak. Similarly, in the array
[−1, −2, −100, 99, 98, 97], the fourth element is the peak.

```python
array =[1,2,3,2,0]
length = len(array)
peak_value = array[0];
max_elemnt = 0; 
for i in range(0,length,1):
    if array[i] > peak_value:
        peak_value = array[i]
        max_elemnt = i+1 
    
print("Highest element is ",max_elemnt)
```

A list of marks obtained by twenty students in an exam is given: [70, 67, 56, 90, <br>
78, 68, 87, 89, 87, 85, 86, 76, 75, 69, 74, 74, 84, 83, 77, 88]. Now, write three <br>
different functions in Python to determine the following: <br>
(a) The mean() to calculate the mean mark. <br>
(b) The maxmin() to find the highest and the lowest mark. <br>
(c) The std\_dev() to calculate the standard deviation. <br>
(d) Repeat the above three steps using NumPy library functions. <br>

```python
import math
import numpy as np
def mean(list):
    length = len(list)
    sum =0 
    for l in list:
        sum += l
    result = sum/length
    return result

def maxmin(list):
    length = len(list)
    maxval = list[0];
    minval = list[0]; 
    for i in range(0,length,1):
        if list[i] > maxval:
            maxval = list[i]
        if list[i] < minval:
            minval = list[i]
    return maxval,minval

def std_dev(list):
    avg = mean(list);
    sum =0;
    for i in list:
        x = (i -avg)**2 
        sum += x
    result = math.sqrt(sum/len(list))
    return result
list =[70, 67, 56, 90, 78, 68, 87, 89, 87, 85, 86, 76, 75, 69, 74, 74, 84, 83, 77, 88]
maximum,minimum = maxmin(list)
arr = np.array(list)
 
print("Mean :" , mean(list))
print("Maximum :" , maximum," Minimum: ",minimum)
print("standard deviation: " ,std_dev(list)) 
print("using Numpy")
print("Mean :" , np.mean(arr))
print("Maximum :" , np.max(arr)," Minimum: ",np.min(arr))
print("standard deviation: " ,np.std(arr)) 

   
```

```python
def matrix_dimensions(test_list):
    # Calculate the number of rows by finding the length of the input list.
    row = len(test_list)
    # Calculate the number of columns by finding the length of the first sub-list (assuming all sub-lists have the same length).
    column = len(test_list[0])
    return row, column  # Return the number of rows and columns as a tuple.

def printMatrix(matrix):
    R,C =matrix_dimensions(matrix)
    for i in range(R):
        for j in range(C):
            print(matrix[i][j], end = " ")
        print()
    print("\n")
```

Write a function in Python that calculates the square of a matrix. The code <br>
should take the following as inputs from the user: <br>
(a) The dimension of the matrix, and <br>
(b) The matrix values <br>

```python
def square_matrix(matrix):
     R,C = matrix_dimensions(matrix)
     
     for i in range(R):
          for j in range(C):
               matrix[i][j] = matrix[i][j]**2
     return matrix
```

```python
def matrix_multiplication(matA,matB):
    r1,c1 = matrix_dimensions(matA)
    r2,c2 = matrix_dimensions(matB)
    output = [[0 for i in range(c2)] for i in range(r1)]
    for i in range(r1):
        for j in range(c2):
            sum = 0
            for k in range(r2):
                sum += matrixA[i][k] * matrixB[k][j]
            output[i][j] = sum
    return output
    

def  MultiplicationPossible(matA,matB):
    a = len(matA[0])
    b = len(matB)
    if a==b:
        return True
    else:
        return False
```
